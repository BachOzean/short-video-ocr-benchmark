```python
import os

video_id_list = df["video_id"].head(5).tolist()

def download_single(video_id):
    local_file = f"tmp/{video_id}.mp4"
    oss_object = f"videos/{video_id}.mp4"
    bucket.get_object_to_file(oss_object, local_file)
    print("downloaded:", local_file)

for vid in video_id_list:
    download_single(vid)
    # os.remove(f"tmp/{vid}.mp4")
    
```
